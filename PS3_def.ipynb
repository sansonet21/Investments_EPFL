## Question 1

We considered an investor who wants to invest in N risky assets with return $R_i$ $\forall i = 1,...,N$ with expected return $\mathbb{E}[R_i] = \mu_i$ and $V[R_i] = \sigma_i^2$. The investor seeks a portfolio which return $R_p = R_f + w^{T} (R - R_f*I)$ maximizes her mean-variance objective function $U(w) =  \mathbb{E}[R_p] - 0.5 * \gamma *  V[R_p] $, where $w$ is the asset portfolio weight vector.

## 1.a

 We know that $\mathbb{E} [R_p] = R_f + w^{T}( \mu - R_f I)$ and that $ V[R_i] = w^{T} \Sigma w$, while we want to minimize $U(w)$.
 To do so, we compute $ \dfrac{\partial f(w)}{\partial w} = \mu - R_f I - \gamma \Sigma w = 0$. <p> Taking the i-th component: 
 $ \mu_i - R_f = \gamma cov(R_i, R_f) $ $\forall i = 1,...,N$ <p>
Since $ (\Sigma w)_i = \sum_{j=1}^{N}{\Sigma_{i,j} * w_j} = \sum_{j=1}^{N}{cov(R_j,R_i) * w_j} = cov( R_i, \sum_{j =1}^{N} w_j R_j = cov(R_i,R_p) $
$\blacksquare$ 

## 1.b
Since it holds for every asset i, it holds also for every portfolio.
Considering $R_i = R_p$, we get: $\mu_p - R_f = \gamma V[R_p]$ 
Then, $ \gamma = \dfrac{\mu_p - R_f}{V[R_p]}$ <p>
Therefore, $\mu_i-R_f = \dfrac{\mu_p - R_f}{V[R_p]} * cov (R_i,R_p) = \beta_{i,p} * (\mu_p- R_f) $<p>
Where $\beta_{i,p} = \dfrac{cov (R_i,R_p)}{V[R_p]} $ is the linear regression coefficinet of $R_i$ on the mean variance efficient portfolio return $R_p$

## 1.c 
we define the linear regression model: 
$R_i = \alpha_i + \beta_{i,p} (R_p - R_f) + \epsilon_i $ <p> with $ \mathbb{E}[\epsilon_i] = 0 $ and $\mathbb{E}[\epsilon_i(R_p-R_f)] = 0 $ <p>
Running the regression we get:
$\beta_{i,p} = \dfrac{cov (R_i,R_p)}{V[R_p]} $ since $R_f$ is a constant <p>
Taking the expected value we get: 
$$\mathbb{E}[R_i] = \alpha_i + \beta_{i,p} \mathbb{E}[R_p - R_f] + \mathbb{E}[\epsilon_i]$$
$$\mu_i = \alpha_i + \beta_{i,p} (\mu_p - R_f)$$
from the previous point, we can conclude that $\alpha_i = R_i$, <p>
since $\mathbb{E}[\epsilon_i] = 0 $ and $\mathbb{E}[\epsilon_i(R_p-R_f] = cov( \epsilon_i, R_p-R_f) = cov(\epsilon_i,R_p) = 0 $


 ## 1.d
 A portfolio on the mean-variance efficient frontier is a linear combination of therisk-free asset and the tangent portfolio. Therefore, two mean-variance efficient portfolios have a correlation equal to 1. <p>
 Thus, 
 $$\dfrac{\mu_p-R_f}{\sigma_p} = \dfrac{(\mu_q - R_f) \dfrac{cov(R_p,R_q)}{\sigma_q^2}}{\sigma_p} = \dfrac{(\mu_q - R_f)\dfrac{\sigma_p \sigma_q \rho }{\sigma_q^2}}{\sigma_p} = \dfrac{\mu_q-R_f}{\sigma_q}$$

## (a)

```python
stocks = db.raw_sql("select date,vwretd from crsp.msi where date>='1960-01-01'", date_cols=['date'])
bonds = db.raw_sql("select caldt,b2ret from crsp.mcti where caldt>='1960-01-01'", date_cols=['caldt'])
stocks.set_index('date',inplace=True)
bonds.set_index('caldt',inplace=True)
```

```python
stock_mean=stocks.values.mean()*12
stock_std=stocks.values.std()*np.sqrt(12)
bond_mean=bonds.values.mean()*12
bond_std=bonds.values.std()*np.sqrt(12)
data=pd.merge(stocks,bonds,left_index=True,right_index=True)
cov=data.cov().values*12
mean_60_40=0.6*stock_mean+0.4*bond_mean
w=np.array([0.6,0.4])
std_60_40= np.sqrt ( w @ cov @ w.T )
rf=0.01
SR_60_40=(mean_60_40-0.01)/std_60_40
print('The mean of the 60-40 portfolio is : ',mean_60_40)
print('The std of the 60-40 portfolio is : ',std_60_40)
print('The sharpe ratio of the 60-40 portfolio is : ',SR_60_40)
```

```python
ones=np.ones((1,2))
mu=np.array([stock_mean,bond_mean])
A=np.dot(ones,np.dot(np.linalg.inv(cov),np.transpose(ones)))
B=np.dot(ones,np.dot(np.linalg.inv(cov),np.transpose(mu)))
C=np.dot(mu,np.dot(np.linalg.inv(cov),np.transpose(mu)))
D=A*C-B**2
w_tan=np.linalg.inv(cov)@(mu-rf*ones).T/(B-A*rf)
print('w_tan : ',w_tan)
```

```python
mean_tan=w_tan.T@mu
std_tan=np.sqrt(w_tan.T @ cov @ w_tan)
SR_tan=(mean_tan-rf)/std_tan
print('The mean of the tangent portfolio is : ',mean_tan[0])
print('The std of the tangent portfolio is : ',std_tan[0,0])
print('The sharpe ratio of the tangent portfolio is : ',SR_tan[0,0])
```

## (b)

```python
K=1/(1/stock_std+1/bond_std)
w_rp=K*np.array([1/stock_std,1/bond_std]).T
mean_rp=w_rp.T@mu
std_rp=np.sqrt(w_rp.T @ cov @ w_rp)
SR_rp=(mean_rp-rf)/std_rp
print('The mean of the tangent portfolio is : ',mean_rp)
print('The std of the tangent portfolio is : ',std_rp)
print('The sharpe ratio of the tangent portfolio is : ',SR_rp)
```

The std of a portfolio that invest $\alpha$ in the risk parity portfolio and $(1-\alpha)$ in the risk-free asset is $ \alpha*std_{rp}$ and it should be equal to the std of the 60-40 portfolio. 

```python
alpha=std_60_40/std_rp
w_lrp=alpha*w_rp
w_lrp_rf=1-ones@w_lrp
mean_lrp=alpha*mean_rp+(1-alpha)*rf
std_lrp=alpha*std_rp
SR_lrp=(mean_rp-rf)/std_lrp
print('w_stocks : ',w_lrp[0],'w_bonds : ',w_lrp[1],'w_rf : ',w_lrp_rf[0])
print('The mean of the tangent portfolio is : ',mean_lrp)
print('The std of the tangent portfolio is : ',std_lrp)
print('The sharpe ratio of the tangent portfolio is : ',SR_lrp)
```

```python
w1=np.linspace(-0.5,1.5,1000)
muw=w1*mu[0]+(1-w1)*mu[1]
stdw=np.sqrt(w1**2*cov[0,0]+2*w1*(1-w1)*cov[1,0]+(1-w1)**2*cov[1,1])
w2=np.linspace(-0.2,3,10000)
mu_rf=w2*mean_tan[0]+(1-w2)*rf
std_rf=abs(std_tan[0]*w2)

plt.plot(std_rf,mu_rf,label='Mean-std deviation frontier with risk-free asset')
plt.plot(stdw,muw,label='Mean-std deviation frontier without risk-free asset')
plt.plot(std_tan,mean_tan,'o',label='Tangent portfolio')
plt.plot(std_lrp,mean_lrp,'o',label='Levered risk-parity portfolio')
plt.plot(std_60_40,mean_60_40,'o',label='60-40 portfolio')
plt.plot(std_rp,mean_rp,'o',label='Risk-parity portfolio')
plt.legend()
plt.show()
```

The difference between levered risk parity and risk parity performance can be explained by the leverage effect. In fact, combining the risk parity portfolio with the risk-free asset it is possible to achieve higher values of expected return. On the other hand the use of leverage increases the standard deviation of the portfolio. The sharpe ratio of the two portfolios is exactly the same. If someone wants to invest part of his wealth in the risk-parity portfolio and the remaining in the risk free asset,the quantity he wants to borrow would depend on his risk aversion.

We know that the general form of the tangent portfolio is:
$$w=\gamma\Sigma^-{1}(\mu-r_f1)$$
And therefore the weights of the risky asset for a mean-variance efficient portfolio would be proportional to these.<p>
If $\Sigma$ is diagonal, i.e. if assets are uncorrelated, the weights of mean-variance efficient portfolio would be proportional to the inverse of the variance.
