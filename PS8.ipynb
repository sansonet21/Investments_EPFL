```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
lam=0.0001
gamma=1 #suppose the risk aversion is 1
mu=0
N=15
#we assume to have 14 trading periods and teherfore the optimal trading strategy 
#would consist of 15 values with the first one equal to 1.000.000
n0=1000000
sigma1=0.3/np.sqrt(252*(N-1)) 
#we assume the 30 min volatility to be like that considering 14 periods per day
Q=np.ones((N,1))
q=np.ones((N,1))
c=np.ones((N,1))
Q[N-1]=lam*gamma*sigma1**2/(lam+gamma*sigma1**2)
q[N-1]=mu*lam/(lam+gamma*sigma1**2)
c[N-1]=0.5*mu**2/(lam+gamma*sigma1**2)
```

```python
for i in range(N-2,0,-1):
    Q[i] = (lam*(Q[i+1]+gamma*sigma1**2)/(Q[i+1]+gamma*sigma1**2+lam))
    q[i] = lam*(q[i+1]+mu)/(Q[i+1]+gamma*sigma1**2+lam)
    c[i] = c[i+1]+0.5*(q[i+1]+mu)**2/(Q[i+1]+gamma*sigma1**2+lam)

aim=q/Q
tau=(Q+gamma*sigma1**2)/(Q+gamma*sigma1**2+lam)
n=np.zeros((N,1))
tran_cost=np.zeros((N,1))
n[0]=n0
for i in range(1,N): 
    n[i]= tau[i]*aim[i]+(1-tau[i])*n[i-1]
    tran_cost[i] = tran_cost[i-1]+0.5*lam*(n[i]-n[i-1])**2
```

```python
plt.plot(n)
plt.ylabel('number of stocks')
plt.xlabel('period')
plt.title('Trades every 30 minutes')
plt.show()
```

```python
plt.plot(tran_cost)
plt.ylabel('transaction costs')
plt.xlabel('period')
plt.title('Trades every 30 minutes')
plt.show()
```
